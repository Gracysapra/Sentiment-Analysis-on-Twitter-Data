```python
# Perform cleaning and preprocessing on the 'text' column
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer

nltk.download('stopwords')
def clean_text(text):
    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove punctuation
    text = re.sub(f'[{string.punctuation}]', '', text)
    # Remove newlines
    text = re.sub(r'\n', '', text)
    # Remove alphanumeric words (words containing digits)
    text = re.sub(r'\b\w*\d\w*\b', '', text)
    # Convert to lowercase
    text = text.lower()
    # Remove remaining non-alphabetic characters (except spaces)
    text = re.sub(r'[^a-z\s]', '', text)
    # Normalize repeated characters (e.g., "soooo" -> "so")
    text = re.sub(r'(.)\1+', r'\1\1', text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = text.split()
    words = [word for word in words if word not in stop_words]

    # Stem words using LancasterStemmer
    stemmer = LancasterStemmer()
    words = [stemmer.stem(word) for word in words]

    # Join words back into a single string
    text = ' '.join(words)

    return text

# Ensure the 'text' column is of string type
df_train['text'] = df_train['text'].astype(str)
data_test['text'] = data_test['text'].astype(str)

# Apply the clean_text function and assign it back to the DataFrame
df_train['text'] = df_train['text'].apply(clean_text)
data_test['text'] = data_test['text'].apply(clean_text)
```

<h3> Synonym Replacement

```python
# Synonyms Replacemnt
import random
from nltk.corpus import wordnet

def synonym_replacement(text, n=1):
    words = text.split()
    new_words = words.copy()
    random_word_list = list(set([word for word in words if wordnet.synsets(word)]))
    random.shuffle(random_word_list)

    num_replaced = 0
    for random_word in random_word_list:
        synonyms = wordnet.synsets(random_word)
        if synonyms:
            synonym = synonyms[0].lemmas()[0].name()
            new_words = [synonym if word == random_word else word for word in new_words]
            num_replaced += 1
        if num_replaced >= n:
            break

    return ' '.join(new_words)

# Apply it on training data
X_train_augmented = [synonym_replacement(text) for text in x_train]
```

<h3> Random Insertion

```python
import random
from nltk.corpus import wordnet

def random_insertion(text, n=1):
    words = text.split()

    # If the text is empty or has less than 2 words, return the text as is
    if len(words) < 2:
        return text

    new_words = words.copy()

    for _ in range(n):
        # Ensure words list is not empty
        if len(words) > 0:
            random_word = random.choice(words)

            # Get synonyms for the random word
            synonyms = wordnet.synsets(random_word)

            # If synonyms are found, insert one synonym randomly
            if synonyms:
                synonym = synonyms[0].lemmas()[0].name()
                random_position = random.randint(0, len(new_words) - 1)
                new_words.insert(random_position, synonym)

    return ' '.join(new_words)

# Apply it on training data
X_train_augmented += [random_insertion(text) for text in x_train if len(text.split()) > 0]  # Ensure non-empty text
```

<h3> Random Deletion

```python
def random_deletion(text, p=0.2):
    words = text.split()
    if len(words) <= 1:  # Check for both empty and single-word cases
        return text

    new_words = [word for word in words if random.uniform(0, 1) > p]

    if len(new_words) == 0:
        # If all words are removed, return the original text instead of choosing from an empty list
        return text
    return ' '.join(new_words)

# Apply it on training data
X_train_augmented += [random_deletion(text) for text in x_train]
```

```python
from googletrans import Translator

translator = Translator()

def back_translation(text):
    try:
        # Translate from English to French and back to English
        translation = translator.translate(text, src='en', dest='fr').text
        return translator.translate(translation, src='fr', dest='en').text
    except Exception as e:
        print(f"Error translating text: {text} - {str(e)}")
        return text  # Return the original text if translation fails
```

```python
X_train_augmented = x_train.tolist()  # Assuming X_train is in sparse format, convert to list
y_train_augmented = y_train.tolist()

# Apply data augmentation on training data
for text, label in zip(x_train, y_train):
    augmented_text = random_insertion(text)  # Augment text
    X_train_augmented.append(augmented_text)  # Add augmented text
    y_train_augmented.append(label)  # Add the corresponding label for the augmented text

# After augmentation, ensure that both X_train_augmented and y_train_augmented have the same length
print(len(X_train_augmented), len(y_train_augmented))  # They should be the same
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the training data
X_train_vectorized = vectorizer.fit_transform(X_train_augmented)

# Transform the testing data
X_test_vectorized = vectorizer.transform(x_test)
```

<h1> Machine Learning (ML) Models </h1>


```python
# Analyze the results
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics

def analyze_results(y_test,y_pred):
    # Get the metrics
    accuracy = metrics.accuracy_score(y_test, y_pred)
    precision =  metrics.precision_score(y_test, y_pred, average='weighted')
    recall = metrics.recall_score(y_test, y_pred, average='weighted')
    f1_score = metrics.f1_score(y_test, y_pred, average='weighted')

    # print metrics
    print("Accuracy:", "%.6f" % accuracy)
    print("Precision:", "%.6f" % precision)
    print("Recall:", "%.6f" %  recall)
    print("F1 Score:", "%.6f" %  f1_score)
    print("Classification Report:")
    print(metrics.classification_report(y_test, y_pred))

    # Produce a confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Neutral', 'Positive', 'Negative'],
                yticklabels=['Neutral', 'Positive', 'Negative'])
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title('Confusion Matrix')
    plt.show()

    return accuracy, precision, recall, f1_score
```

<h2> Logistic Regression </h3>

```python
# prompt: apply logistic regression

from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression model
logistic_regression = LogisticRegression(max_iter=1000)

# Train the model
logistic_regression.fit(X_train_vectorized, y_train_augmented)

# Make predictions on the testing data
y_pred_lr = logistic_regression.predict(X_test_vectorized)

# Analyze the results for Logistic Regression
accuracy_lr, precision_lr, recall_lr, f1_score_lr = analyze_results(y_test, y_pred_lr)
```

<h2> Random Forest Classifier</h2>

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest Classifier
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
random_forest.fit(X_train_vectorized, y_train_augmented)

# Make predictions on the testing data
y_pred_rf = random_forest.predict(X_test_vectorized)

# Analyze the results for Random Forest
accuracy_rf, precision_rf, recall_rf, f1_score_rf = analyze_results(y_test, y_pred_rf)
```

<h2> Support Vector Machine (SVM) Classifier </h2>

```python
from sklearn.svm import SVC

# Initialize the SVM model
svm_classifier = SVC(kernel='linear', random_state=42)

# Train the model
svm_classifier.fit(X_train_vectorized, y_train_augmented)

# Make predictions on the testing data
y_pred_svm = svm_classifier.predict(X_test_vectorized)

# Analyze the results for SVM
accuracy_svm, precision_svm, recall_svm, f1_score_svm = analyze_results(y_test, y_pred_svm)
```

<h2> Ensemble Learning - AdaBoost Algorithm </h2>

```python
from sklearn.ensemble import AdaBoostClassifier

# Initialize the AdaBoost Classifier with a base estimator (e.g., Decision Tree)
adaboost_classifier = AdaBoostClassifier(n_estimators=50, random_state=42)

# Train the model
adaboost_classifier.fit(X_train_vectorized, y_train_augmented)

# Make predictions on the testing data
y_pred_adaboost = adaboost_classifier.predict(X_test_vectorized)

# Analyze the results for AdaBoost
accuracy_adaboost, precision_adaboost, recall_adaboost, f1_score_adaboost = analyze_results(y_test, y_pred_adaboost)
```
